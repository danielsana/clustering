```python
# Regression and Classification are Supervised Learning  - There is model testing
# Clustering - is Unsupervised Learning  - No model testing.
# Clustering Gruping data to different groups called clusters
# Each of These clusters has similarities, something in common.
# Applications: Customer Segmentation, Data Segmentation

import pandas as pd
data = pd.read_csv("Mall_Customers.csv")
data.head()
```

```python
# check empties, data wragling 
data.isnull().sum()
```

```python
data.describe()
```

```python
# fill empties in Gender column 
data['Gender'].fillna('unknown', inplace=True)
```

```python
# fill empties in Annual Income column
# calculate the mean 
mean = data['Annual Income'].mean()
data['Annual Income'].fillna(mean, inplace=True)
```

```python
# fill empties in Spending Score column
# calculate the mean 
mean = data['Spending Score'].mean()
data['Spending Score'].fillna(mean, inplace=True)
```

```python
# check empties, data wragling 
data.isnull().sum()
```

```python
array = data.values
X = array[:, 2:5]   # Age, Income, Score
# No Variable Y
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=10, random_state=42)
model.fit(X)
```

```python
# Lets see the 10 Clusters Kmeans created.
means = model.cluster_centers_
clusters = pd.DataFrame(means, columns = ['Age', 'Income', 'Score'])
clusters
```

```python
# Let see who is in group 2
data['no']  = model.labels_   # get every record group number
cluster_old =  data[data['no'] == 0]
cluster_old
```

```python
data[data['no'] == 0].to_csv('Cluster0.csv')
```
